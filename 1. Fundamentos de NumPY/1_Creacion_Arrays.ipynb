```python
# Arrays de ceros y unos

array_ceros = np.zeros((3, 3))
array_unos = np.ones((2,2))
```

```python
print(f"Array de ceros: \n {array_ceros}")
print(f"Array de unos: \n {array_unos}")
```

```python
# Arrays de rangos

array_rango = np.arange(10)
```

```python
print(f"Array de rango: \n {array_rango}")
```

```python
# Arrays espaciados linealmente

array_espaciado = np.linspace(0, 1, num=5)
```

```python
print(f"Array espaciado linealmente: \n {array_espaciado}")
```

```python
# Matriz identidad

m_identidad = np.eye(4)
```

```python
print(f"Matriz identidad: \n {m_identidad}")
```

```python
# Arrays con tipos de datos especificos

int_array = np.array([[1, 2], [3, 4]], dtype='int32')
float_array = np.array([[1, 2], [3, 4]], dtype='float32')
```

```python
print(f"Array con valores enteros: \n {int_array}")
print(f"Array con valores flotantes: \n {float_array}")
```
