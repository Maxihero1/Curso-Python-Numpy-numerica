```python
import numpy as np
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
```

```python
# Suma de arrays

print(a + b)
```

```python
# Resta de arrays

print(a - b)
```

```python
# Escalar de arrays

print(a + 10)
print(a - 10)
print(a * 10)
print(a / 10)
```

```python
# Multiplicacion de arrays

print(a * b)
```

```python
# Division de arrays

print(a / b)
```
