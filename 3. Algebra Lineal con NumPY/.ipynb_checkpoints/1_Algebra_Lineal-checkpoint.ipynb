```python
import numpy as np
```

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[2, 0], [1, 2]])
```

```python
# Multiplicacion de matrices

# Primera manera
C = np.dot(A, B)
print(C)

# Segunda manera
D = A @ B
print(D)
```

```python
# Multiplicacion de vectores
X = np.array([2, 3])
Y = np.array([5, 2])

Z = X @ Y
print(Z)
```

```python
# Determinante de una matriz
np.linalg.det(A)
```

# Importante: Regla a considerar de una matriz inversa.
Si la determinante de una matriz es 0, esta matriz no tiene inversa.

```python
# Inversa de una matriz
det_a = np.linalg.det(A)
if det_a != 0:
    inv_A = np.linalg.inv(A)
    print(f"Inversa de la matriz A:\n {inv_A}")
else:
    print("La matriz A no tiene inversa")
```

```python
# Ejemplo de una matriz sin inversa
M = np.array([[0, 1], [0, 1]])
det_M = np.linalg.det(M)

if det_M != 0:
    inv_M = np.linalg.inv(M)
    print(f"Inversia de la matriz M:/n {inv_M}")
else:
    print("La matriz M no tiene inversa.")
```

```python
# Descomposicion de una matriz QR
Q, R = np.linalg.qr(A)

print(f"Factor Q:\n {Q}")
print(f"Factor R:\n {R}")
```
