# ¿Por que usar funciones en lugar de usar operaciones directamente?
Si bien es cierto que "**a+b**" es equivalente a "**np.add(a, b)**" cuando hacemos uso de la funcion tenemos varias posibilidades.

Una de estas por ejemplo es cambiar el tipo de dato del array, es decir, si usamos "**np.add(a, b, dtype='int32')**" cambiamos el tipo del todos los elementos resultantes de dicha operación a **int32**. Esto puede ser util en situaciones donde se requiera dicha accion (Ya sea por ahorro de espacio, etc).

Existen otros parametros que pueden ser adicionados en dichas funciones bajo otros escenarios.

```python
import numpy as np
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
```

```python
# Suma de arrays
np.add(a, b)
```

```python
# Resta de arrays
np.subtract(a, b)
```

```python
# Multiplicacion de arrays
np.multiply(a, b)
```

```python
# Division de arrays
np.divide(a, b)
```

```python
# Nota: en esta variable los valores pertenecientes al vector son considerados angulos.
angulos = np.array([0, 30, 60, 90])
```

```python
# Funcion de seno
np.sin(angulos)
```

```python
# Nota: en esta variable pasamos los valores de angulos a radianes.
angulos_rad = np.radians(angulos)
```

```python
np.sin(angulos_rad)
```

```python
# Funcion de coseno
np.cos(angulos_rad)
```

```python
# Funcion de tangente
np.tan(angulos_rad)
```

```python
# Operaciones logicas
np.greater(a, b) # Compara si los elementos del primer array del parametro son mayores a los elementos del segundo.
np.less(a, b) # Compara si los elementos del primer array del parametro son menores a los elementos del segundo.
np.equal(a, b) # Compara si los elementos del primer array del parametro son iguales a los elementos del segundo.
```

```python
# Funciones exponenciales
np.power(a, b) # Eleva los elementos del primer array con los elementos del segundo array
np.mean(a) # Da el promedio del array
np.std(a) # Calcula la desviacion estandar del array
```
