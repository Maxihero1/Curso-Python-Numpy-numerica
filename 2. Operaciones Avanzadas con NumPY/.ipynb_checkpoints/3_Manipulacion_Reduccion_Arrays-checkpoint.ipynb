```python
import numpy as np
```

```python
# Haciendo uso del metodo copy() garantizo que la manipulacion de la variable principal no afecte a otras variables que hagan uso de esta.
# El metodo copy() hace que use una copia de la variable en lugar de la variable original. Hacer esto implica utilizar mas memoria
z = np.array([2, 3, 4])
x = z.copy()

z[1] = 9
```

```python
print('X: ', x)
print('Z: ', z)
```

```python
a = np.arange(1, 10)
print(a)

# Reshape modifica el array a una matriz.
b = a.reshape((3, 3))
print(b)
```

```python
# Si quiero aplanar una matriz puedo usar flatten() y ravel().
c = b.flatten()
print("Metodo flatten()")
print("Variable C previo a la modificacion: ", c)
c[0] = 99
print("Variable B posterior a la modificacion: \n", b)
print("Variable C posterior a la modificacion: ", c, "\n")

d = b.ravel()
print("Metodo ravel()")
print("Variable D previo a la modificacion: ", d)
d[0] = 99
print("Variable B posterior a la modificacion: \n", b)
print("Variable D posterior a la modificacion: ", d)
```

# ¿Que diferencia el metodo flatten() del metodo ravel()?
Cuando hacemos uso del metodo flatten() la variable que va a ser aplanado deja de tener relacion con la variable original. En cambio con el metodo ravel() la variable que va a ser aplanado sigue teniendo relacion con la variable original.

En pocas palabras, si hago "**c = b.flatten()**", y posterior a esto modifico la variable c, digase "**c[0] = 99**" solamente cambiaria la variable c. En cambio, si uso "**c = b.ravel()**", y posteriormente modifico la variable c, digase c[0] = 99, tanto la variable c como la variable b se verían afectados.

Entonces, se puede concluir con que el metodo flatten() hace uso de una "vista" de la variable a modificar, mientras que ravel() hace uso directo de la variable a modificar. El alcance de ambos metodos no es el mismo.

```python
# Transposicion de matrices (metodo transpose() y metodo T)
print(b.transpose())

print(b.T)
```

```python
# Suma de los elementos de una matriz
print(np.sum(b))
```

```python
# Elemento maximo de una matriz
print(np.max(b))
```

```python
# Elemento minimo de una matriz
print(np.min(b))
```

```python
# Promedio entre los elementos de una matriz
print(np.mean(b))
```

```python
# Desviacion estandar de una matriz
print(np.std(b))
```

```python
# Metodos aplicados a columnas y filas especificas de una matriz
# Si deseo el promedio de las columnas axis = 0, y de las filas axis = 1
print(np.mean(b, axis = 0))
print(np.std(b, axis = 0))
print(np.sum(b, axis = 0))
print(np.max(b, axis = 0))
print(np.min(b, axis = 0))
```
